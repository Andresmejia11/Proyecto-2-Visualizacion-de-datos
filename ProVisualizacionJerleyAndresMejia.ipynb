```python
import geopandas as gpd
import pandas as pd
import re

# Cargar archivo de ubicaciones
facilities = gpd.read_file('/content/drive/MyDrive/Colab Notebooks/Visualizacion/facilities.gpkg')

# Cargar archivos CSV
minerals = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Visualizacion/minerals.csv')
capacity = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Visualizacion/capacity.csv')
reserves = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Visualizacion/reserves.csv')

# Verificar columnas
print(facilities.columns)
print(minerals.columns)
print(capacity.columns)
print(reserves.columns)
minerals.head(5)
```

```python
facilities_sub = facilities[['facility_id', 'facility_name', 'facility_type',
                             'primary_commodity', 'country', 'geometry']]
minerals_sub = minerals[['facility_id', 'material', 'value_tonnes']]
# material = tipo de mineral, "value_tonnes" = producción
capacity_sub = capacity[['facility_id', 'value_tpa']]
# value_tpa = capacidad anual
reserves_sub = reserves[['facility_id', 'mineral_value_tonnes']]
# mineral_value_tonnes = reservas disponibles
#facilities_sub.columns
```

```python
# Merge para crear un solo dataframe
df = facilities_sub.merge(minerals_sub, on='facility_id', how='left')
df = df.merge(capacity_sub, on='facility_id', how='left')
df = df.merge(reserves_sub, on='facility_id', how='left')

# Guardar CSV final global
df.to_csv('/content/drive/MyDrive/Colab Notebooks/Visualizacion/mining_final.csv', index=False)

```

```python
# Cargar CSV union europea final
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Visualizacion/mining_final_UE.csv')

# Para uso en flourish se es necesario extrar del dato MULTIPOINT la longitud y latitud
# Función para extraer lon y lat desde
def extract_lon_lat(geom):
    if isinstance(geom, str):
        # Buscamos dos números decimales que pueden tener signo negativo
        # y que están separados por un espacio.
        match = re.search(r"(-?\d+\.\d+)\s+(-?\d+\.\d+)", geom)
        if match:
            return float(match.group(1)), float(match.group(2))
    return None, None

# Aplicar extracción
df[['longitude', 'latitude']] = df['geometry'].apply(
    lambda x: pd.Series(extract_lon_lat(x))
)

# Guardar nuevo CSV listo para Flourish
df.to_csv(
    '/content/drive/MyDrive/Colab Notebooks/Visualizacion/mining_final_flourish.csv',
    index=False
)

print("CSV listo para Flourish")
print(df[['geometry', 'longitude', 'latitude']].head())
```
